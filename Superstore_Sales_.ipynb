# **Clustering for customer segmentation**

To segment customers or products based on key features using clustering and visualize the clusters to understand the characteristics of each segment

```python
# Choose relevant features for clustering. In this cases, lets use Sales, Profit, and Quantity to segment customers or producrs based on their purchasing beavior.
#Select relevant features
clustering_data = df[['Sales', 'Profit', 'Quantity', 'Discount']]

# Basic scaling of features
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
clustering_data_scaled = scaler.fit_transform(clustering_data)

# Apply K-means Clustering:
# Use the K-means algorithm to segment customers or products into distinct clusters
from sklearn.cluster import KMeans

# Define number of clusters
kmeans = KMeans(n_clusters=3, random_state=42)
Clusters = kmeans.fit_predict(clustering_data_scaled)

# Add cluster labels to the original data
df['Cluster'] = Clusters

df.head()
```

```python
# Visualize the cluster - Use Plotly to create a 3D scatter plot to better visualize clusters across Sales, Profit, and Quantity.
import plotly.express as px

fig = px.scatter_3d(df, x='Sales', y='Profit', z='Quantity', color='Cluster',
                    title='K-Means Clustering of Customers')
fig.show()
```

Cluster 0 - low-value customers: low sales volume and low profitability.

Cluster 1 - high sales and profit, it may represent high-value customers

Cluser 2 - high frequency and low profitability, it may represent frequent customers of low ticket items.

```python
# Calculate average sales, profit, and quantity for each cluster to understand the business value of each segment.
# Calulate average metrics for each cluster
cluster_summary =df.groupby('Cluster').agg({'Sales': 'mean', 'Profit': 'mean', 'Quantity': 'mean', 'Discount': 'mean'}).reset_index()
print(cluster_summary)
```

```python
# Compare clusters with existing customer segments
comparison_table = pd.crosstab(df['Segment'], df['Cluster'])
print(comparison_table)
```

Cluster 0 - Dominated by consumers, might represent customers with moderate purchasing behavior

Cluster 1 - small, diverse group. They may make occasional, high-value purchases

Cluster 2 - leans toward high-value Consumer and Corporate customers

```python
# Use Plotly to create a bar chart that shows the average Sales, Profit, and quantity for each cluster.

import plotly.express as px

#Bar chart for cluster characteristics
fig = px.bar(cluster_summary, x='Cluster', y=['Sales', 'Profit', 'Quantity'],
             title='Cluster Characteristics',
             labels=['Sales', 'Average Value', 'Variable' 'Metric'],
             barmode='group')

fig.show()
```

# Regression Modeling

```python
df.info()
```

```python
# Regression Modeling
from sklearn.linear_model import LinearRegression

# Reset the index to restore the 'Order Date' column
df = df.reset_index()

#Convert 'Order Date to DateTime format
#df_reg['Order Date'] = pd.to_datetime(df['Order Date'])

# Extract month and year from 'Order Date'
df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month
```

```python
#Convert categorical variables into numerical form using one-hot encoding
df = pd.get_dummies(df, columns=['Category', 'Segment'], drop_first=True)
print(df.head())
```

```python
#Select indepedent variables features and dependent variables target for regession model

X = df[['Discount', 'Quantity', 'Sales','Year', 'Month',
            'Category_Office Supplies', 'Category_Technology']]
y = df['Profit']

#Add Sales * Discount to account for how discounts impact high_priced items differently.
#X['Sales_Discount'] = X['Sales']*X['Discount']

#Split the data into a training set for model training and a testing set for evaluation.
from sklearn.model_selection import train_test_split

#Split data: 70% for training, 30% for testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
#Train a linear regression model
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
#Get the coefficents and intercept

coefficents = pd.DataFrame({'Feature': X.columns, 'Coefficients': model.coef_})
intercept = model.intercept_

#Display coefficients
print("intercept:", intercept)
print(coefficents)

# The coefiicents represent the change in profit for one unit increase in each variabele,
# holding others constant. for example, if sales has a high positive coefficient, increasing sales should increase profit
```

Discount reduces profit. The superstore might want to consider implementing a different promotion strategy.

Increasing quantity decreases profit. This could indicate low-margin products or icreased costs.

Office Supplies and Technology are the most profitable categories.

```python
#Evaluate the model
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
import numpy as np

#Make predictions on the test set
y_pred = model.predict(X_test)

#Calculate metrics
r2 = r2_score(y_test, y_test)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)

#Print metrics
print('R-squared (r2):', r2)
print('Mean Absolute Error (MAE):', mae)
print('Mean Squared Error (MSE):', mse)
print('Root Mean Squared Error (RMSE):', rmse)
```

```python
# Visualize prediction vs actual profit
# A scatter plot helps me visually assess how well the model's predictions align with actual profit values
import matplotlib.pyplot as plt

#Plot predicted vs actual profit
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred, alpha=0.5,color='blue')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color='red', linestyle = "--") # line for perfect prediction
plt.xlabel('Actual Profit')
plt.ylabel('Predicted Profit')
plt.title('Prediction vs Actual Profit')
plt.show()
```

# Classification

```python
# Generate a target variable: High vs low profitability
# Create a new column 'Total Profit' by grouping by customer
customer_profit = df.groupby('Customer ID')['Profit'].sum().reset_index()
print(customer_profit)
customer_profit['profitability'] = customer_profit['Profit'].apply(lambda x: 1 if x > customer_profit['Profit'].mean() else 0)

#Merge back to original dataset
df = df.merge(customer_profit[['Customer ID', 'profitability']], on='Customer ID')
df.head()
```

High-profitability customers are categorized as 1.

```python
# Train decision tree model
from sklearn.tree import DecisionTreeClassifier
Classifier = DecisionTreeClassifier(max_depth=5, random_state=42)
Classifier.fit(X_train, y_train)

# Make predictions
y_pred = Classifier.predict(X_test)

from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
print('Classification Report:\n', classification_report(y_test, y_pred))
print('Accuracy:', accuracy_score(y_test, y_pred))
print('Confusion Matrix:\n', confusion_matrix(y_test, y_pred))
```

```python
import plotly.express as px

# Create a column to indicate if the prediction is correct
visualization_df = X_test.copy()
visualization_df['True Profitability'] = y_test
visualization_df['Predicted Profitability'] = y_pred
visualization_df['Correct Prediction'] = (
    visualization_df['True Profitability'] == visualization_df['Predicted Profitability']
)

# Map correctness to color for clarity
fig = px.scatter(
    visualization_df,
    x='Sales',
    y='Quantity',
    color='Correct Prediction',  # True or False for correct predictions
    symbol='True Profitability',  # Shape by the true label
    title='Classification Results: True vs Predicted Profitability',
    labels={'Correct Prediction': 'Prediction Correct?'},
    color_discrete_map={True: "green", False: "red"}  # Green for correct, red for incorrect
)

# Display the plot
fig.show()
```
